```python
import requests

# Your NewsAPI key
API_KEY = "ad2f08a3a6fc4765be2fdea9544f6cdc"

# Function to fetch news for a given topic
def get_news(topic):
    url = f"https://newsapi.org/v2/everything?q={topic}&sortBy=publishedAt&pageSize=10&apiKey={API_KEY}"
    response = requests.get(url)
    if response.status_code == 200:
        data = response.json()
        articles = data.get("articles", [])
        return articles
    else:
        print(f"Failed to fetch news for {topic}: {response.status_code}")
        return []

# Function to display news articles
def display_news(topic, articles):
    print(f"\nTop 10 Articles for {topic.capitalize()}:\n" + "="*50)
    for i, article in enumerate(articles, start=1):
        print(f"{i}. {article['title']}")
        print(f"   Link: {article['url']}\n")

# Topics to fetch news for
topics = [
    "cryptocurrency",
    "stock market",
    "economy",
    "politics",
    "world news",
    "tech and AI",
    "emerging technologies"
]

# Fetch and display news for each topic
if __name__ == "__main__":
    for topic in topics:
        articles = get_news(topic)
        display_news(topic, articles)
```
